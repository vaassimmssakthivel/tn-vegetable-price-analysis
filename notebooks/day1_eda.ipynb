## FIRST REAL ANALYSIS

# Average prices by district

```python
df.groupby('district')['retail_price_inr_per_kg'].mean()
```

# Sorts districts:highest price → lowest price.

```python
df.groupby('district')['retail_price_inr_per_kg'].mean().sort_values(ascending=False)
```

## SEASONAL ANALYSIS

# Which season has highest average retail price?

```python
df.groupby('season')['retail_price_inr_per_kg'].mean()
```

## TRANSPORT COST ANALYSIS

# Which transport mode is most expensive?

```python
df.groupby('transport_mode')['transport_cost_inr'].mean()
```

## CREATE FIRST CHART

```python
import matplotlib.pyplot as plt

district_price = (
    df.groupby('district')['retail_price_inr_per_kg']
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

district_price.plot(kind='bar')

plt.title("Average Retail Price by District")

plt.xlabel("District")

plt.ylabel("Retail Price")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig(
    "../images/district_price_chart.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

# Key Findings

1. Chennai has higher average retail prices.
2. Summer season shows increased pricing.
3. Certain transport modes increase costs significantly.

# Final Business Insights

1. Chennai and Coimbatore exhibit consistently higher retail prices.

2. Summer season contributes to increased post-harvest losses.

3. Certain transport modes significantly increase supply-chain costs.

4. Tomato and onion prices show high volatility across seasons.

5. Margin percentages vary substantially between districts.
