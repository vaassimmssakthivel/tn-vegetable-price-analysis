```python
import pandas as pd

df = pd.read_csv(
    "../data/TN_Vegetable_Price_Index_Enhanced.csv"
)

df = df.drop_duplicates()

df['date'] = pd.to_datetime(df['date'])

df['profit_margin'] = (
    df['retail_price_inr_per_kg']
    - df['farm_price_inr_per_kg']
)

df.to_csv(
    "../data/cleaned_vegetable_data.csv",
    index=False
)
```
